### La compression du Fichier PDF avec Ghostscript

```python
import os
import fitz 
import subprocess

# Définir explicitement le chemin de Ghostscript pour Windows
gs_path = r"C:\Program Files\gs\gs10.04.0\bin\gswin64c.exe"  #Reécrit le chemin exact selon l'installation

# Optimiser un fichier PDF
input_pdf_path = "mon_fichier.pdf"
output_pdf_path = "mon_fichier_optimise.pdf"

# Commande pour exécuter la compression via Ghostscript
gs_command = [
    gs_path, "-sDEVICE=pdfwrite", "-dCompatibilityLevel=1.4",
    "-dPDFSETTINGS=/screen", "-dNOPAUSE", "-dQUIET", "-dBATCH",
    f"-sOutputFile={output_pdf_path}", input_pdf_path
]

# Exécuter la commande
subprocess.run(gs_command, check=True)
print("Compression terminée.")
```
